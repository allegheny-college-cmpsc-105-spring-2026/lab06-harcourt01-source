```python
# load pre-existing tools available for data exploration

import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import interact # pre existing code for interaction
```

```python
# data_analytics.csv

# update the column names to match the names you have in your csv file for data_analytics

# create a pandas dataframe
dfa = pd.read_csv(
    '/content/data_analytics.csv',
    sep=',',
    names = ['Page', 'Views', 'Active users', 'Views per active user',
             'Average engagement time per active user', 'Event count', 'Key events', 'Total revenue'],
    skiprows=10
)

# Display the first few rows of the DataFrame
print(dfa.head())
```

```python
# data_survey.csv

# update the column names to match the names you have in your csv file for data_survey

# create a pandas dataframe
dfs = pd.read_csv(
    '/content/data_survey.csv',
    sep=',',
    names = ['Timestamp', 'Have you played a sport?', 'What sport did you play?',
             'How long did you play your sport?', 'How would you rate your sport?'],
    skiprows=1
)

# Display the first few rows of the DataFrame
print(dfs.head())
```

```python
# how many rows and columns are in the dataframe dfa?
print(f'Shape: {dfa.shape}')
# how many rows and columns are in the dataframe dfs?
print(f'Shape: {dfs.shape}')
```

```python
# Analytics

# Do one or more explorations with dfa
# no interaction is required

# make a blank canvas
plt.figure()

# make a title starting with 'Figure 1:' Fill in a relevant title
plt.title('Figure 1: Views vs Active Users')

# plot
plt.bar(dfa['Page'], dfa['Views'])

# ylabel
plt.ylabel('Views')

# xlabel
plt.xlabel('Page')

# ylim
plt.ylim(0, 10)

# reveal everything
plt.show()
```

```python
# Survey

# Do one or more explorations with dfs
# no interaction is required
dfs_clean = dfs.dropna(subset=['What sport did you play?'])

# make a blank canvas
plt.figure(figsize= (9,4))
# make a title starting with 'Figure 2:' Fill in a relevant title
plt.title('Figure 2: Sport Ratings')

plt.bar(dfs_clean['What sport did you play?'], dfs_clean['How would you rate your sport?'])# plot
plt.ylabel('')# ylabel
# xlabel
# ylim

plt.show()# reveal everything
```
